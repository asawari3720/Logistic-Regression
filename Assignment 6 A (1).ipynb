```python
bankdata.drop(["pdays"],inplace=True,axis=1)
bankdata.drop(["month"],inplace=True,axis=1)
bankdata.drop(["education"],inplace=True,axis=1)
bankdata.head(2)
```

```python
y1= bankdata["y"]
x1= bankdata.drop("y",axis=True)
classifier = LogisticRegression()
classifier.fit(x1,y1) 
```

```python
classifier.coef_
```

```python
classifier.predict_proba (x1)
```

```python
y_pred = classifier.predict(x1)
bankdata["y_pred"] = y_pred
bankdata
```

```python
y_prob = pd.DataFrame(classifier.predict_proba(x1.iloc[:,:]))
new_df = pd.concat([bankdata,y_prob],axis=1)
new_df
```
